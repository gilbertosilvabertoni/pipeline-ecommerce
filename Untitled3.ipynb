```python
import requests
import csv
```

```python
url = "https://fakestoreapi.com/products"
response = requests.get(url)

if response.status_code != 200:
    raise Exception(f"Falha na requisição: {response.status_code}")
```

```python
products = response.json()
```

```python
with open("products.csv", "w", encoding="utf-8", newline="") as file:
    writer = csv.writer(file)

    writer.writerow([
      "id",
      "title",
      "price",
      "category",
      "rating_rate",
      "rating_count"
    ])

    for product in products:
        # Extrai os campos básicos
        product_id = product["id"]
        title = product["title"]
        price = product["price"]
        category = product["category"]

        # O campo `rating` vem como um objeto: { "rate": x, "count": y }
        # Aqui pegamos `rate` e `count` e "achatamos" em duas colunas
        rating = product["rating"]
        rating_rate = rating["rate"]
        rating_count = rating["count"]

        # Escreve uma linha no CSV com todos os campos
        writer.writerow([
            product_id,
            title,
            price,
            category,
            rating_rate,
            rating_count
        ])
```

```python
from google.colab import files
uploaded = files.upload()

import pandas as pd

df = pd.read_csv("products.csv")
print(df.head())
print(df.shape)
```

```python
# Remove espaços sobrando no começo e fim dos títulos
df["title"] = df["title"].str.strip()

# Coloca a categoria em maiúsculo
df["category"] = df["category"].str.upper()

print(df[["title", "category"]].head())
```

```python
# Converte o preço de dólar pra real
COTACAO = 5.70

df["price_brl"] = df["price"] * COTACAO

# Arredonda pra 2 casas decimais
df["price_brl"] = df["price_brl"].round(2)

print(df[["title", "price", "price_brl"]].head())
```

```python
# Função que classifica o produto pelo preço em reais
def classificar_tier(preco):
    if preco <= 100:
        return "budget"
    elif preco <= 500:
        return "mid"
    else:
        return "premium"

# Aplica a função em cada linha da coluna price_brl
df["price_tier"] = df["price_brl"].apply(classificar_tier)

print(df[["title", "price_brl", "price_tier"]].head(10))
```

```python
# Remove a coluna price original (em dólar)
df = df.drop(columns=["price"])

# Garante a ordem das colunas conforme solicitado pelo cliente
df = df[["id", "title", "price_brl", "category", "rating_rate", "rating_count", "price_tier"]]

# Salva o arquivo final
df.to_csv("products_transformed.csv", index=False)

# Faz o download no seu computador
from google.colab import files
files.download("products_transformed.csv")

print(f"Arquivo salvo com {len(df)} linhas e {len(df.columns)} colunas.")
print(df.head())
```
